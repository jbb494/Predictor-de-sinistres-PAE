```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_csv('../OnlineData/DataToCorrelate.csv')
```

```python
# print(data.columns)
tipusSinistre = ['AccidenteDeFumador',
       'AccidenteDomestico', 'ActosVandalicos', 'AsentamientosHundimientos',
       'AveriaDeLaInstalacion', 'AveriaInternaDelAparato', 'CaidaDeObjetos',
                 'Cortocircuito', 'ElectricidadEstatica', 'Emboces',
       'Expolacion', 'CaidaDeRayo', 'Filtraciones',
       'FugasInstalacionesDeExtincion', 'GolpeDeAire', 'GolpeFortuito',
       'GoterasProcedentesDeLocales', 'Humo', 'Hurto', 'IgnicionEspontanea',
       'ImpactoVehiculos', 'IncendiosDeOrigenElectrico', 'IntentoDeRobo',
       'InundacionDesbordamientoDesviacion', 'LlamasAbiertas', 'Lluvia',
       'AveriasAparatosElectricos', 'ManipulacionMaquinariaBienesAsegurados',
       'MaterialesRecalentados', 'MontajeAnomalo',
       'Nieve', 'NinoJugando', 'OlvidoCierreGrifos', 'OndasSonicas', 'Otros',
       'Pedrisco', 'PicoTensionCiaElectrica',
       'ProvenienteFuegoExteriorPorPropagacion', 'RCNoImputableAlAsegurado',
       'RCDanosMateriales', 'Robo',
       'RocesFriccionesChispasMecanicas', 'RoturasPorObras',
       'RoturaTuberiaAlimentacion', 'RoturaTuberiaCalefaccion',
       'RoturaTuberiaDesague', 'SoldaduraYCorte',
       'SuperficiesCalientes', 'TormentaElectrica', 'Viento']


dadesACorrelacionar = ['EdatConstruccion', 'Clase_Alta', 'Clase_MediaAlta', 'Clase_Media',
       'Clase_MediaSencilla', 'Clase_Sencilla', 'Temperatura_max',
       'Temperatura_min', 'Humitat_max', 'Humitat_min', 'Precipitacio_max',
       'Precipitacio_min', 'IrrSolar_max', 'IrrSolar_min', 'Vent_max',
       'Vent_min', 'ProbabilitatManifestacio', 'ProbabilitatFesta',
       'VolumSinistres']
print(tipusSinistre)
print(dadesACorrelacionar)
```

```python
dictCorr = {}
clf = DecisionTreeClassifier(criterion="entropy", random_state=0)
```

```python

doTree = True
first = True
TreeVariables = ["Lluvia", "Filtraciones", "Viento", "TormentaElectrica", "RoturaTuberiaAlimentacion"]
for sinistre in TreeVariables if doTree  else tipusSinistre:
    dictCorr[sinistre] = []
    col1 = data[sinistre].astype(float)
    col4 = pd.DataFrame(col1)
    for variable in dadesACorrelacionar:
        col2 = data[variable].astype(float)
        dictCorr[sinistre].append(col1.corr(col2))
        if doTree:
            if first:
                col3 = pd.DataFrame(col2)
                first = False
            else:
                col3 = col3.join(col2)
    if doTree:
        first = True
        array = cross_val_score(clf, col3, col4, cv=10)
        print(sinistre)
        print(array)
        print('-------------------')
```

```python
dict_sorted = sorted(dictCorr.items(), key=lambda x: max(x[1]), reverse=True)

for key, value in dict_sorted:
    print(key)
    value = [abs(float(i)) for i in value]
    for elem in sorted(value, reverse=True):
        print(str(dadesACorrelacionar[value.index(elem)]) + ': ' + str(elem))
    print('-------------------------------')
```

```python
num = 1

wantedKeys = ['Lluvia', 'Filtraciones', 'RoturaTuberiaAlimentacion', 'TormentaElectrica', 'Viento']
for key, value in dictCorr.items():
    if key in wantedKeys:
        with open(f'DeepLearning/MultiModels/Metadata/{num}.json', 'w') as f:
            f.write('{\n')
            f.write(f'"LabelName": "{key}",\n')
            f.write(f'"InputColsNames": [\n')
            value = [abs(float(i)) for i in value]
            first = True
            for elem in sorted(value, reverse=True):
                if first:
                    f.write(f'"{str(dadesACorrelacionar[value.index(elem)])}"')
                    first = False
                else:
                    f.write(f',\n"{str(dadesACorrelacionar[value.index(elem)])}"')
            f.write('\n]\n}\n')
        num += 1
```
