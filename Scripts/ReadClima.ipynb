```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
import datetime
```

```python
clima = pd.read_csv('../../Dades_meteorol_giques_de_la_XEMA.csv')
```

```python
clima['DATA_LECTURA'] = pd.to_datetime(clima['DATA_LECTURA']).dt.date
codiToCP = { 'D4': [17480], 'X4': [8001], 'D5': [8028], 'X8': [8007], 'X2': [8003], 'D9': [43700], 'XE': [43006],
             'VX': [8593], 'UB': [17134], 'ZD': [17003], 'W9': [17176, 17177, 17178, 17179], 'WT': [8380],
             'V4': [8585], 'CY': [8529] }
valuesDays = clima['DATA_LECTURA'].unique()
```

```python
numCP = 0
for codi_est,CPs in codiToCP.items():
    numCP += len(CPs)
print(numCP)
```

```python
#data = {'Temperatura_max': np.nan,'Temperatura_min': np.nan,'Humitat_max': np.nan,'Humitat_min': np.nan,'Precipitacio_max': np.nan, 'Precipitacio_min': np.nan,'IrrSolar_max': np.nan,'IrrSolar_min': np.nan,'Neu_max': np.nan,'Neu_min': np.nan,'Vent_max': np.nan,'Vent_min': np.nan}
# data = pd.DataFrame(aux)
columns = ['Temperatura_max','Temperatura_min','Humitat_max','Humitat_min','Precipitacio_max', 'Precipitacio_min','IrrSolar_max','IrrSolar_min','Neu_max','Neu_min','Vent_max','Vent_min']
allCols = columns
allCols.append('DATA_LECTURA')
firstAllCols = True
daylist = pd.date_range(start="2017-01-01",end="2019-12-31").to_pydatetime().tolist()
cleanData = pd.DataFrame(index=range(len(daylist*numCP)), columns=columns)
i = 0
for dia in daylist:
    dia = dia.date()
    # en principi assumeixo que les dades de XEMA esta per cada dia
    subarray = clima.loc[ clima['DATA_LECTURA'] == dia ]
    for codi_est,CPs in codiToCP.items():
        #for ind,_ in data.items():
         #   data[ind] = np.nan
        cleanData.at[i,'DATA_LECTURA'] = dia
        auxarray = subarray.loc[ subarray['CODI_ESTACIO'] == codi_est ]
        for index, row in auxarray.iterrows():
            valLectura = np.float64(row['VALOR_LECTURA'])
            if row['CODI_VARIABLE'] == 32: # Temperatura
                if np.isnan(cleanData.at[i,'Temperatura_max']):
                    cleanData.at[i,'Temperatura_max'] = valLectura
                    #cleanData.at[i,'Temperatura_max'] = cleanData.at[i,'Temperatura_max'].astype(float)
                    cleanData.at[i,'Temperatura_min'] = valLectura
                else:
                    cleanData.at[i,'Temperatura_max'] = max(valLectura, cleanData.at[i,'Temperatura_max'])
                    #cleanData.at[i,'Temperatura_max'] = cleanData.at[i,'Temperatura_max'].astype(float)
                    cleanData.at[i,'Temperatura_min'] = min(valLectura, cleanData.at[i,'Temperatura_min'])
            elif row['CODI_VARIABLE'] == 33: # Humitat relativa
                if np.isnan(cleanData.at[i,'Humitat_max']):
                    cleanData.at[i,'Humitat_max'] = valLectura
                    cleanData.at[i,'Humitat_min'] = valLectura
                else:
                    cleanData.at[i,'Humitat_max'] = max(valLectura, cleanData.at[i,'Humitat_max'])
                    cleanData.at[i,'Humitat_min'] = min(valLectura, cleanData.at[i,'Humitat_min'])
            elif row['CODI_VARIABLE'] == 35: # Precipitació
                if np.isnan(cleanData.at[i,'Precipitacio_max']):
                    cleanData.at[i,'Precipitacio_max'] = valLectura
                    cleanData.at[i,'Precipitacio_min'] = valLectura
                else:
                    cleanData.at[i,'Precipitacio_max'] = max(valLectura, cleanData.at[i,'Precipitacio_max'])
                    cleanData.at[i,'Precipitacio_min'] = min(valLectura, cleanData.at[i,'Precipitacio_min'])
            elif row['CODI_VARIABLE'] == 36: # Irradiància
                if np.isnan(cleanData.at[i,'IrrSolar_max']):
                    cleanData.at[i,'IrrSolar_max'] = valLectura
                    cleanData.at[i,'IrrSolar_min'] = valLectura
                else:
                    cleanData.at[i,'IrrSolar_max'] = max(valLectura, cleanData.at[i,'IrrSolar_max'])
                    cleanData.at[i,'IrrSolar_min'] = min(valLectura, cleanData.at[i,'IrrSolar_min'])
            elif row['CODI_VARIABLE'] == 38: # neu
                if np.isnan(cleanData.at[i,'Neu_max']):
                    cleanData.at[i,'Neu_max'] = valLectura
                    cleanData.at[i,'Neu_min'] = valLectura
                else:
                    cleanData.at[i,'Neu_max'] = max(valLectura, cleanData.at[i,'Neu_max'])
                    cleanData.at[i,'Neu_min'] = min(valLectura, cleanData.at[i,'Neu_min'])
            elif row['CODI_VARIABLE'] == 50: # vent
                if np.isnan(cleanData.at[i,'Vent_max']):
                    cleanData.at[i,'Vent_max'] = valLectura
                    cleanData.at[i,'Vent_min'] = valLectura
                else:
                    cleanData.at[i,'Vent_max'] = max(valLectura, cleanData.at[i,'Vent_max'])
                    cleanData.at[i,'Vent_min'] = min(valLectura, cleanData.at[i,'Vent_min'])
        auxIndex = i
        for CP in CPs:
            j = 0
            for valueRow in cleanData.loc[i, :]:
                cleanData.loc[auxIndex, allCols[j]] = valueRow
                j += 1
            cleanData.at[auxIndex,'CP'] = np.int64(CP)
            if firstAllCols:
                allCols.append('CP')
                firstAllCols = False
            auxIndex += 1
        i = auxIndex
    print(dia)
```

```python
# cost logarítmic de l'script Ω(n*m) on n=num dies i m=num codis postals

cleanData.to_csv('../../clean_clima.csv', index=False)
```
